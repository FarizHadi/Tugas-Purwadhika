```python
#Q1
def factorial(num):
    if num == 0 or num == 1:
        return 1
    elif num < 0:
        return None
    else:
        return num * factorial(num - 1)

assert factorial(0) == 1
assert factorial(3) == 6
assert factorial(10) == 3628800
assert factorial(-10) == None
```

```python
#Q2

def timeConvert(totalSecond):
    l1 = []

    hour = totalSecond // 3600
    remain_seconds = totalSecond % 3600

    minute = remain_seconds // 60
    seconds = remain_seconds % 60

    if hour == 0 and minute == 0 and seconds == 0:
        l1.append(f"{seconds}s")
    else:    
        if hour > 0 :
            l1.append(f"{hour}h")
        if minute > 0:
            l1.append(f"{minute}m")
        if seconds > 0:
            l1.append(f"{seconds}s")

    result = ' '.join(l1)
    return result    

assert timeConvert(0) == '0s'
assert timeConvert(30) == '30s'
assert timeConvert(60) == '1m'
assert timeConvert(90) == '1m 30s'
assert timeConvert(3600) == '1h'
assert timeConvert(3601) == '1h 1s'
assert timeConvert(3661) == '1h 1m 1s'
assert timeConvert(90042) == '25h 42s'
```

```python
#Q3
def area(l,w):
    return l * w
def volume(l,w,h):
    return l * w * h

assert area(10, 10) == 100
assert area(0, 9999) == 0
assert volume(10, 10, 10) == 1000
assert volume(5, 8, 10) == 400
```

```python
#Q4
def getColor(col,row):
    if row > 7 or col > 7:
        return ''
    elif row % 2 != 0 and col % 2 != 0: # ganjil ganjil
        return "white"
    elif row % 2 != 0 and col % 2 == 0:  # ganjil genap   
        return "black" 
    elif row % 2 == 0 and col % 2 == 0: # genap genap
        return "white"
    elif row % 2 == 0 and col % 2 != 0: # genap ganjil 
        return "black" 
    else: 
        return "white"       

assert getColor(1, 1) == 'white'
assert getColor(2, 1) == 'black'
assert getColor(1, 2) == 'black'
assert getColor(7, 7) == 'white'
assert getColor(0, 8) == ''
assert getColor(2, 9) == ''
```

```python
#Q6 
def isLeapYear(year):
    if year % 4 == 0:
        # leap year
        if year % 100 == 0:
            #not leap year
            if year % 400 == 0:
                #leap year
                return True
            else:
                return False
        else:
            return True    
    else:
        return False    
    
assert isLeapYear(1999) == False
assert isLeapYear(2000) == True
assert isLeapYear(2001) == False
assert isLeapYear(2004) == True
assert isLeapYear(2100) == False
assert isLeapYear(2400) == True
```

```python
#Q7
def isValidDate(year, month, day):
    if month < 1 or month > 12:
        return False
    if day < 1:
        return False

    if month in [1,3,5,7,8,10,12]: # bulan dengan tgl  max 31
        return day <= 31
    if month in [4,6,9,11]: # bulan dengan tgl max 30
        return day <= 30
    if isLeapYear(year): # bulan februari and check apakah termasuk "leap year"
        return day <= 29
    else:
        return day <= 28


assert isValidDate(2001, 2, 29) == False
assert isValidDate(2029, 13, 1) == False
assert isValidDate(1000000, 1, 1) == True
assert isValidDate(2015, 4, 31) == False
```

```python
#Q8
def commaFormat(number):
    # Split the number into integer and fractional parts
    parts = str(number).split('.')

    # Format the integer part with commas
    integer_part = parts[0]

    formatted_integer = ""

    while len(integer_part) > 3:
        formatted_integer = "," + integer_part[-3:] + formatted_integer
        integer_part = integer_part[:-3]

    formatted_integer = integer_part + formatted_integer

    # Combine the integer and fractional parts, if any
    if len(parts) == 2:
        return formatted_integer + "." + parts[1]
    else:
        return formatted_integer

assert commaFormat(1) == '1'
assert commaFormat(10) == '10'
assert commaFormat(100) == '100'
assert commaFormat(1000) == '1,000'
assert commaFormat(10000) == '10,000'
assert commaFormat(100000) == '100,000'
assert commaFormat(1000000) == '1,000,000'
assert commaFormat(1234567890) == '1,234,567,890'
assert commaFormat(1000.123456) == '1,000.123456' 
```
